## 3.0 Identify the adopter user

```python
# Code task 13#
# Extract date from timestamp for unique login tracking
user_engagement["date"] = user_engagement["time_stamp"].dt.date

# Code task 14#
# Sort values for correct rolling window application
user_engagement = user_engagement.sort_values(by=["user_id", "date"])

# Code task 15#
# Function to check if a user has logged in 3 times within a 7-day window
def is_adopted(user_dates):
    user_dates = sorted(user_dates.unique())  # Ensure unique dates and sorted order
    return any((user_dates[i + 2] - user_dates[i]).days <= 7 for i in range(len(user_dates) - 2))

# Code task 16#
# Group by user_id and check for adoption criteria
adopted_users = user_engagement.groupby("user_id")["date"].apply(is_adopted)

# Code task 17#
# Convert adopted users to a DataFrame
adopted_users_df = adopted_users.reset_index()

# Code task 18#
# Filter for adopted users
adopted_users_df = adopted_users_df[adopted_users_df['date'] == True]

# Code task 19#
# Rename the 'date' column to 'adopted'
adopted_users_df.rename(columns={'date': 'adopted'}, inplace=True)

# Code task 20#
# Display the first few rows of the adopted users DataFrame
adopted_users_df.head()
```

## 4.0 Factors predict future user adoption

```python
# Code task 21#
# Merge adopted users data with the users dataset
users = users.rename(columns={"object_id": "user_id"})  # Ensure correct key for merging

# Code task 22#
# Merge to add adoption status to the user data
users_adopted = users.merge(adopted_users_df, on="user_id", how="left")

# Code task 23#
# Fill missing values in 'adopted' column with 0 (indicating non-adopted users)
users_adopted["adopted"] = users_adopted["adopted"].fillna(0).astype(int)

# Code task 24#
# Display the first few rows of the merged dataset
users_adopted.head()
```

```python
# Code task 25#
# Convert categorical variables into numerical format for correlation analysis
users_adopted_encoded = users_adopted.copy()
users_adopted_encoded = pd.get_dummies(users_adopted_encoded, columns=["creation_source"], drop_first=True)

# Code task 26#
# Select only numerical columns for correlation analysis
numeric_columns = users_adopted_encoded.select_dtypes(include=[np.number])

# Code task 27#
# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Code task 28#
# Extract correlations with the 'adopted' column
correlation_with_adoption = correlation_matrix["adopted"].sort_values(ascending=False)

# Code task 29#
# Display the correlations
print(correlation_with_adoption)
```

### 4.1 Summary

The correlation analysis highlights which factors are most related to user adoption:

. Last Session Creation Time (0.39)

    - This is the strongest positive correlation.
    - Users who logged in recently are much more likely to be adopted users.

. Organization ID (0.067)
    - Users in certain organizations are slightly more likely to be adopted.
    - This suggests some organizations have higher engagement levels.
        
. Marketing Opt-in (0.0088) and Marketing Drip (0.0066)
    - Users who opt into marketing emails or are included in marketing drip campaigns show a very weak correlation with adoption.
    - This suggests marketing campaigns alone are not a strong predictor of user adoption.
